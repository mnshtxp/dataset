```
# tenure 를 20, 40, 60 으로 분할
def tenure_split(df) : 
    if df["tenure"] <=20:
        return "0-20"
    elif df["tenure"] > 20 and df["tenure"] <= 40:
        return "20-40"
    elif df["tenure"] > 40 and df["tenure"] <= 60:
        return "40-60"
    else:
        return "60plus"
        
        
df['tenure_group'] = df.apply(tenure_split, axis = 1)
df[['tenure', 'tenure_group']].sample(10)
```

```
# monthlycharges_split 을 30, 70, 100  범위로 나눕니다.
def monthlycharges_split(df) :   
    if df['MonthlyCharges'] <= 30 :
        return '0-30'
    elif (df['MonthlyCharges'] > 30) & (df['MonthlyCharges'] <= 70 ):
        return '30-70'
    elif (df['MonthlyCharges'] > 70) & (df['MonthlyCharges'] <= 99 ):
        return '70-99'
    elif df['MonthlyCharges'] > 99 :
        return '99plus'
    
df['monthlycharges_group'] = df.apply(monthlycharges_split, axis = 1)
df[['MonthlyCharges', 'monthlycharges_group']].head()
```

```
# TotalCharges 를 2000, 4000, 6000 으로 분할
def totalcharges_split(df) :   
    if df['TotalCharges'] <= 2000 :
        return '0-2k'
    elif df['TotalCharges'] > 2000 and df['TotalCharges'] <= 4000:
        return '2k-4k'
    elif df['TotalCharges'] > 4000 and df['TotalCharges'] <= 6000:
        return '4k-6k'
    else:
        return '6kplus'
    

df['totalcharges_group'] = df.apply(totalcharges_split, axis = 1)
df[['TotalCharges', 'totalcharges_group']].sample(10)
```

## 데이터 분할

```
target = 'Churn'
```

```
# train / validation / test 데이터셋 분리
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size = 0.20, 
                              stratify = df[target], random_state = 2)

train, val = train_test_split(train, test_size = 0.25, 
                              stratify = train[target], random_state = 2)


train.shape, val.shape, test.shape
```

## 정보 누수(leakage) 확인 

```
for item in [train, val, test]:
  item[target] = (item[target] == 'Yes')
```

```
from category_encoders import OrdinalEncoder
from category_encoders import OneHotEncoder

from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```
# feature Matrix와 target vector 나누기
features = train.columns.drop([target])

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
y_test = test[target]

print("feature Matrix: ", X_train.shape, X_val.shape, X_test.shape)
print("target vector: ", y_train.shape, y_val.shape, y_test.shape)
```

```
# Decision
# pipeline 생성
pipe = make_pipeline(
    OrdinalEncoder(), 
    DecisionTreeClassifier(max_depth = 5, random_state=2)
)
pipe.fit(X_train, y_train)

# 예측
y_pred = pipe.predict(X_val)

print("검증 정확도: ", pipe.score(X_val, y_val))
print("검증 f1_score: ", f1_score(y_val, y_pred))
```

```
# Decision OneHot
# pipeline 생성
pipe = make_pipeline(
    OneHotEncoder(), 
    DecisionTreeClassifier(max_depth = 5, random_state=2)
)
pipe.fit(X_train, y_train)

# 예측
y_pred = pipe.predict(X_val)

print("검증 정확도: ", pipe.score(X_val, y_val))
print("검증 f1_score: ", f1_score(y_val, y_pred))
```

클래스의 비율과 비슷하게 검증 정확도가 나왔다.

```
# SHAP를 수행하는 함수 생성
def use_shap(row_number):

  # 선택한 row 가져오기
  row = X_test.iloc[[row_number]]
  row_processed = X_test_encoded.iloc[[row_number]]

  # 예측
  pred = model.predict(row_processed)[0]
  if pred == 1:
    result = '"취소"'
  elif pred == 0:
    result = '"유지"'

  # 해당 클래스에 속할 확률
  pred_proba = model.predict_proba(row_processed)[0, pred]
  pred_proba *= 100

  # 결과 출력
  print(f'이 대출에 대한 예측결과는 {pred} 으로, 확률은 {pred_proba:.0f}% 입니다.')
  print()

  # SHAP
  explainer = shap.TreeExplainer(model)
  shap_values = explainer.shap_values(row_processed)

  # 클래스가 1로 분류되는 것에 영향을 준 정도 확인
  feature_names = row.columns
  feature_value = row.values[0]
  shaps = pd.Series(shap_values[0], zip(feature_names, feature_value))

  # positive 상위 3개
  pros = shaps.sort_values(ascending=False)[:3].index
  
  # negative 상위 3개
  cons = shaps.sort_values(ascending=True)[:3].index

  # 예측에 가장 영향을 많이 준 상위 3개 출력

  if pred == 1:
    positive_features = pros
    negative_features = cons
  elif pred == 0:
    positive_features = cons
    negative_features = pros

  print('이 예약 건이 {}로 예측되는 것에 영향을 준 요인 상위 3개'.format(result))
  for idx, info in enumerate(positive_features, start=1):
    idx_name, idx_value = info
    print('{}. {}: {}'.format(idx, idx_name, idx_value))
  print()

  print('이 예약 건이 {}로 예측되는 것에 부정적인 영향을 준 요인 3개'.format(result))
  for idx, info in enumerate(negative_features, start=1):
    idx_name, idx_value = info
    print('{}. {}: {}'.format(idx, idx_name, idx_value))

  # SHAP 출력
  shap.initjs()
  return shap.force_plot(
      base_value = explainer.expected_value,
      shap_values=shap_values,
      features=row,
      link='logit'
)
```
